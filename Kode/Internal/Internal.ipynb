```python
import pandas as pd
import numpy as np

data = pd.read_csv("dataset_internal.csv")
data
```

```python
title = data['title']
title
```

```python
title.to_numpy().shape
```

```python
title = np.array(title, dtype=str)
```

# Preprocessing

```python
title = np.char.lower(title)
```

```python
import re
marks = "!\"#$%&'()*+,-./:;<=>?@[\]^_`{|}~\n"
 
for x in marks:  
    title = np.char.replace(title, x, "")
```

```python
title = repr(title)
title = re.sub(r"[^a-zA-Z]", " ", title)
```

```python
# remove the whitespaces or specific characters from the string at the beginning and end of the string
title = title.strip()
```

```python
title = re.sub(r"\s+", " ", title)
```

```python
from nltk.tokenize import word_tokenize 

title_tokens = word_tokenize(title)
print(title_tokens)
```

```python
from nltk.corpus import stopwords
list_stopwords = stopwords.words('indonesian')
list_stopwords
```

```python
#remove stopword pada list token
def stopwords_removal(words):
    return [word for word in words if word not in list_stopwords]

title = stopwords_removal(title_tokens)

print(title)
```

# TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

max_features = 500

# Feature Engineering 
print ("------- TF-IDF on data -------")

tf_idf = TfidfVectorizer(max_features=max_features, binary=True)
tfidf_mat = tf_idf.fit_transform(title).toarray()

print("TF-IDF ", type(tfidf_mat), tfidf_mat.shape)
```

```python
terms = tf_idf.get_feature_names()

# sum tfidf frequency of each term through documents
sums = tfidf_mat.sum(axis=0)

# connecting term to its sums frequency
data = []
for col, term in enumerate(terms):
    data.append((term, sums[col] ))

ranking = pd.DataFrame(data, columns=['term','rank'])
ranking.sort_values('rank', ascending=False)
```

```python
import gensim
import gensim.corpora as corpora

# Membuat Dictionary
dictionary = corpora.Dictionary([title])

# Membuat Corpus
titles = title

# Term Document Frequency
corpus = [dictionary.doc2bow(title) for text in titles]

# print
print(corpus[:1][0][:30])
```

# LDA

```python
from pprint import pprint
# number of topics
num_topics = 10

# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=dictionary,
                                       num_topics=num_topics)
# Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```
